#### 1. Pg. 115 #2(a) : $f^\prime(1) \space where \space f(x) = x^4 âˆ’ 8x.$

```python
##defining function that will allow us to zoom in on f(1)##
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    #calculating yprime for each x.
    return x ** 4 - 8 * x

def plot(xinit, xend, steps):
    #generate x values.
    x_range = np.linspace(xinit, xend, steps)
    
    #generate y values.
    y = func(x_range)
    #calculate and print slope.
    print('Slope: ', (y[-1]-y[0]) / (x_range[-1] - x_range[0]))
    
    #plot the coordinates.
    plt.plot(x_range, y)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)
    plt.show()
```

```python
plot(0, 2, 100)
```

```python
plot(0.75, 1.25,100)
```

```python
plot(0.96, 1.04, 100)
```

```python
plot(0.97, 1.03, 100)
```

```python
#2a. plotting f(x) = 1 / x
##defining function that will allow us to zoom in on f(2)##
import numpy as np
import matplotlib.pyplot as plt

def func2(x):
    #calculating yprime for each x.
    return 1 / x

def plot2(xinit, xend, steps):
    #generate x values.
    x_range = np.linspace(xinit, xend, steps)
    
    #generate y values.
    y = func2(x_range)
    #calculate and print slope.
    print('Slope: ', (y[-1]-y[0]) / (x_range[-1] - x_range[0]))
    
    #plot the coordinates.
    plt.plot(x_range, y)
    
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)
    plt.show()
```

```python
plot2(1, 3, 100)
```

```python
plot2(1.999999, 2.000001, 100)
```

```python
#2a. defining function to calculate derivative with formula f(a + h) - f(a - h) / 2h where x = a. f(x) = 1 / x
def deriv(x):
    h = [0.01, 0.001, 0.0001, 0.00001, 0.000001]
    for i in range(len(h)):
        d = ((1 / (x + h[i])) - (1 / (x - h[i]))) / (2 * h[i])
        print(d)
    return
deriv(2)
```

The linear approximation formula allows us to substitute the values provided for $f(a)$ and $f^\prime (a)$. in order to solve the problem.

$f(a) = b$ <br>
$f^\prime(a) = -3$ <br>
$f(a+k)\approx f(a)+f^\prime (a) \times k$ **linear approximation formula** <br>
$f(a+k)\approx b +(-3) \times k$ **Substitute** <br>
$f(a+k)\approx b - 3k$ **Simplify** <br>

This is an informal proof that shows that, out of the choices given, $b-3k$ will best approximate $f(a+k)$.

```python
import matplotlib.pyplot as plt
#plotting f(x) = x^3 with each calculation.
def num8(l, u):
    a = 1
    #h = 0.01
    h_list = []
    q1_list = []
    q2_list = []
    q3_list = []
    q4_list = []
    for k in range(0, 9):
        h = 1 / 2 ** k
        q1 = ((a + h) ** 3 - (a - h) ** 3) / h
        q2 = ((a + h) ** 3 - (h ** 3)) / h
        q3 = ((a + h) ** 3 - (a - h) ** 3) / (2 * h)
        q4 = ((a + h) ** 3 - (a - h) ** 3) / (3 * h)

        h_list.append(h)
        q1_list.append(q1)
        q2_list.append(q2)
        q3_list.append(q3)
        q4_list.append(q4)

    #print(q1, q2, q3, q4)
    
    plt.plot(h_list, q1_list, marker = "o", label = "f(a + h) - f(a - h) / h")
    plt.plot(h_list, q2_list, marker = "o", label = "f(a + h) - f(h) / h")
    plt.plot(h_list, q3_list, marker = "o", label = "f(a + h) - f(a - h) / 2h")
    plt.plot(h_list, q4_list, marker = "o", label = "f(a + h) - f(a - h) / 3h")
    plt.grid(True)
    plt.xlabel('h')
    plt.ylabel('deriv')
    plt.ylim(l, u)
    plt.legend()

    return
```
