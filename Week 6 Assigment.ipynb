```python
#Activity 2. TABLE from page 266.

import math
#function for cos(t^2). Working the example from the book.
def fun(t):
    answer = math.cos(t**2)
    return answer
#function for table.
def table(tinit, tfinal, numsteps):
    #time delta calculation, as we've seen in SIRPLOT.
    delta_t = (tfinal - tinit) / numsteps
    t = tinit
    #accumulations always start at 0
    accum = 0

    for k in range(0, numsteps):
        #calculating delta y.
        delta_y = fun(t) * delta_t
        # new accumulation - current y + delta_y
        accum = accum + delta_y
        #increment t.
        t = t + delta_t
        # print delta y, current accumulation, and t. Rounded to 4 digits, to match the table for cos(t^2) in the book.
        print("delta y: ", round(delta_y, 4), " ", "accumulation: ", round(accum, 4), " ", "t: ", t)
#initial t = 0, final t = 4, numsteps = 2^3 = 8.
table(0, 4, 8)
```

Explanation: The accumulation function shows the accumulated change for y over time.  Each step reflects the addition of the current y to the change in y, as represented by the variable delta_y in the code above.  Delta_y is calculated by multiplying the input function of t, $cos (t^2)$ in this case, by delta_t.  We have seen some of this before.  For example, the delta_t calculation is the same as what was used in SIRVALUE.  In order to find the exact value of the accumulation of delta_y, we would need to perform the calculations many more times than what appears above.  

```python
#Activity 3. PLOT from page 167. 
import matplotlib.pyplot as plt
#helper function, allows for user entry of the formula.  I thought that this would be a better way of showing the data for the formulas
#on pg 4. a-f.
def fun2():
    function = input("Enter formula:")
    return lambda t: eval(function)
#plot function, as defined in the text.
def plot(tinit, tfinal, numsteps):
    #time delta calculation, as we've seen in SIRPLOT.
    delta_t = (tfinal - tinit) / numsteps
    t = tinit
    accum = 0
    #declaring our function (user provided).
    f = fun2()
    #lists for x axis (time) and y axis (accumulation of delta y)
    time = []
    acc = []
    for k in range(0, numsteps):
        delta_y = f(t) * delta_t
        accum = accum + delta_y
        t = t + delta_t
        time.append(t)
        acc.append(accum)
    plt.plot(time, acc)
    plt.grid(True)
    plt.xlabel("Time")
    plt.ylabel("Accumulation")
    plt.show()
```

Explanation: The PLOT function is a sumple way to plot our accumulations of delta y on a graph that are a result of the function TABLE, above.  This is similar to the function SIRPLOT that we used to plot SIRVALUE earlier in the semester.  The code above allows for the user to input a function to be used in the delta_y calculation.  PLOT is a great way to observe the value of increasing the number of steps in order to plot a smooth curve.  If we decrease the number of steps, our plot is not as accurate.  An example of this idea can be found for part A below.  As we've seen in our work with SIRPLOT, the same logic applies here.  More frequent analyses result in a more accurate representation of the data.
