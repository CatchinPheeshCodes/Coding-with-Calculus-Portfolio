#### Activity 1: Code Riemann

```python
import numpy as np
import matplotlib.pyplot as plt

def fun(): 
    function = input("Enter your function:")
    return lambda x: eval(function)

def riemann(start, stop, numsteps):
    a = start
    b = stop
    steps = numsteps
    delta_x = (b - a) / steps
    f = fun()
    endpoint = input("Choose left, right or midpoint:")
    plot = input("Enter y if you want to plot, n otherwise:")
    table = input("Enter y if you want a table, n otherwise:")
    accum = 0
    delta_s = 0

    if endpoint == "left":
        x = a
        x_values = np.linspace(a, b - delta_x, steps)
    elif endpoint == "right":
        x = a + delta_x
        x_values = np.linspace(a + delta_x, b, steps)
    elif endpoint == "midpoint":
        x = (a + delta_x) / 2
        x_values = np.linspace(a + delta_x / 2, b - delta_x / 2, steps)
        
    if table == "y":
        
        for k in range(steps):
            delta_s = f(x) * delta_x
            accum = accum + delta_s
            x = x + delta_x
            print(f"x: {x - delta_x}  delta_s: {delta_s}  Accumulation: {accum}")
            
    if table == "n":
        for k in range(steps):
            delta_s = f(x) * delta_x
            accum = accum + delta_s
            x = x + delta_x
        
        print(f"Final Accumulation: {accum}")

    if plot == "y":
        heights = f(x_values)
        areas = heights * delta_x
        total_area = np.sum(areas)
    
        x_plot = np.linspace(a, b, 1000)
        y_plot = f(x_plot)
        
        plt.figure(figsize=(10, 6))
        plt.plot(x_plot, y_plot, 'b-', label='f(x)')
        
        for x, h in zip(x_values, heights):
            plt.bar(x, h, width=delta_x, align='edge', edgecolor='black', alpha=0.5, color='lightblue')
        
        plt.title(f'Riemann Sums')
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.legend()
        plt.grid(True)
        plt.show()
        
```

#### Activity 4 - Riemann with trapezoidal or simpsons rule.

```python
def fun(): 
    function = input("Enter your function:")
    return lambda x: eval(function)

def simpsons(start, stop, numsteps):
    a = start
    b = stop
    steps = numsteps
    delta_x = (b - a) / steps
    f = fun()
    accum = 0
    delta_s = 0


    left_x = a
    right_x = a + delta_x
    midpoint_x = (a + delta_x) / 2
    simp = (1/6) * (4 * midpoint_x + left_x + right_x)

    
    for k in range(steps):
        delta_s = f(simp) * delta_x
        accum = accum + delta_s
        simp = simp + delta_x
        print(f"x: {simp - delta_x}  delta_s: {delta_s}  Accumulation: {accum}")
            
  
```

```python
simpsons(1, 3, 4)
```
